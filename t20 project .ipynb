```python
X=pd.DataFrame(df[['Mat','Inns']])
Y=pd.DataFrame(df['Runs'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
XTR,XT,YTR,YT=train_test_split(X,Y,train_size=0.8, random_state=1)
```

```python
model=LinearRegression()
model.fit(XTR,YTR)
```

```python
#linear regression
```

```python
Model=LinearRegression()
Model= Model.fit(XTR,YTR)
Predict=Model.predict(XT)
RMSE = np.sqrt(mean_squared_error(YT,Predict))
MAPE=mean_absolute_percentage_error(YT,Predict)
print(RMSE, MAPE)
```

```python
#voting Regressor
```

```python
base_model=[('lr',LinearRegression()),('dt',DecisionTreeRegressor()),('rf',RandomForestRegressor())]
```

```python
model= VotingRegressor(estimators=base_model)
model.fit(XTR, YTR)
Predict=model.predict(XT)
RMSE = np.sqrt(mean_squared_error(YT,Predict))
MAPE=mean_absolute_percentage_error(YT,Predict)
print(RMSE, MAPE)
```

```python
#Bagging Regressor
```

```python
BModel=BaggingRegressor(estimator=Model, n_estimators=50, random_state=1)
BModel= BModel.fit(XTR,YTR)
Predict=BModel.predict(XT)
RMSE = np.sqrt(mean_squared_error(YT,Predict))
MAPE=mean_absolute_percentage_error(YT,Predict)
print(RMSE, MAPE)
```

```python
#Stacking
```

```python
base_models =[('DT', DecisionTreeRegressor(random_state=1)),
('RF', RandomForestRegressor(random_state=1)),
('LR', LinearRegression())]
```

```python
meta_model = LinearRegression()
```

```python
stack_model = StackingRegressor(estimators=base_models, final_estimator=meta_model)
```

```python
stack_model
```

```python
Model = stack_model.fit(XTR, YTR)
```

```python
#This project aimed to predict a player's **Total Runs** using two key features: **Matches (Mat)** and **Innings (Inns)**. We evaluated five regression models using **RMSE** (Root Mean Squared Error) and **MAPE** (Mean Absolute Percentage Error).
#The **Decision Tree Regressor** is the best-performing model based on the Mean Absolute Percentage Error (MAPE).
#Lowest Proportional Error: The Decision Tree Regressor achieved the lowest MAPE (**0.281**), indicating that, on average, its prediction error was the smallest proportion of the actual number of runs.
#RMSE Trade-Off: It's important to note that while the Decision Tree had the best MAPE, it also had the **highest RMSE (216.90)**. This suggests the model is volatile: it's often proportionally accurate, but when it is wrong, the absolute error (in terms of actual runs) is significantly larger than the errors made by the Stacking Regressor.
#Recommended Model (Overall):For a stable, production-ready model that minimizes the magnitude of severe errors, the **Stacking Regressor** (RMSE **176.67**) is typically preferred, even though its MAPE is slightly higher.
```
